```python
import numpy as np
from sklearn import model_selection
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
```

```python
X, y = load_iris(return_X_y=True)

# Для наглядности возьмем только первые два признака (всего в датасете их 4)
X = X[:, :2]
```

```python
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
cmap = ListedColormap(['red', 'green', 'blue'])
plt.figure(figsize=(7, 7))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=cmap)
```

```python
def e_metrics(x1, x2):
    
    distance = 0
    for i in range(len(x1)):
        distance += np.square(x1[i] - x2[i])
                
     
    return np.sqrt(distance)
```

```python
def knn(x_train, y_train, x_test, k, weight, q = 0.5):
    
    answers = []
    
    for x in x_test:
        test_distances = []
            
        for i in range(len(x_train)):
            
            # расчет расстояния от классифицируемого объекта до
            # объекта обучающей выборки
            distance = e_metrics(x, x_train[i])
            
            # Записываем в список значение расстояния и ответа на объекте обучающей выборки
            test_distances.append((distance, y_train[i]))
        
        # создаем словарь со всеми возможными классами
        classes = {class_item: 0 for class_item in set(y_train)}
        
        # Сортируем список и среди первых k элементов подсчитаем частоту появления разных классов
        if weight == 'number':
            for d in sorted(test_distances)[0:k]:
                # Рассчитываем вес в зависимости от номера соседа
                w = q**d[0]
                classes[d[1]] += w 
        elif weight == 'distance':
            for d in sorted(test_distances)[0:k]:
                # Рассчитываем вес в зависимости от расстояния до соседа
                w = 1/np.sqrt(d[0])
                classes[d[1]] += w
        # Записываем в список ответов наиболее часто встречающийся класс
        answers.append(sorted(classes, key=classes.get)[-1])
    return answers
```

```python
def accuracy(pred, y):
    return (sum(pred == y) / len(y))
```

```python
def get_graph(X_train, y_train, k, weight):
    cmap_light = ListedColormap(['#FFAAAA', '#AAFFAA','#00AAFF'])

    h = 0.1

    # Расчет пределов графика
    x_min, x_max = X_train[:, 0].min() - 1, X_train[:, 0].max() + 1
    y_min, y_max = X_train[:, 1].min() - 1, X_train[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

    # Получим предсказания для всех точек
    Z = knn(X_train, y_train, np.c_[xx.ravel(), yy.ravel()], k, weight)

    # Построим график
    Z = np.array(Z).reshape(xx.shape)
    plt.figure(figsize=(7,7))
    plt.pcolormesh(xx, yy, Z, cmap=cmap_light, shading='auto')

    # Добавим на график обучающую выборку
    plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap=cmap)
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())
    plt.title(f"Трехклассовая kNN классификация при k = {k}")
    plt.show()
```

```python
k_number = []
# значение k для метрики с добавлением веса в зависимости от номера соседа
for k in range(1, 21):

    y_pred = knn(X_train, y_train, X_test, k, weight = 'number')
    print(f'Точность алгоритма при k = {k}: {accuracy(y_pred, y_test):.3f}')
    get_graph(X_train, y_train, k, weight = 'number')
    k_number.append(accuracy(y_pred, y_test))
```

```python
k_distance = []
# значение k для метрики с добавлением веса в зависимости от расстояния до соседа
for k in range(1, 21):

    y_pred = knn(X_train, y_train, X_test, k, weight = 'distance')
    print(f'Точность алгоритма при k = {k}: {accuracy(y_pred, y_test):.3f}')
    get_graph(X_train, y_train, k, weight = 'distance')
    k_distance.append(accuracy(y_pred, y_test))
```

```python
plt.plot(k_number, label='k_number')
plt.plot(k_distance, label='k_distance')
plt.xlabel('Количество кластеров')
plt.ylabel('Точность алгоритма')
plt.legend(loc='upper right')
plt.show()
```

Алгоритм с метрикой с добавлением веса в зависимости от номера соседа показывает более высокий результат, чем алгоритм с метрикой с добавлением веса в зависимости от расстояния до соседа (0.867 против 0.833)

Оптимальное значение k для метрики с добавлением веса в зависимости от номера соседа: 5. Точность алгоритма 0.867
